```python
import pandas as pd
df=pd.read_csv('Dataset_sensors.csv')
df
```

```python
import matplotlib.pyplot as plt

# Plotting histogram
plt.hist(df['solar_radiation'], bins=20, color='blue', edgecolor='black')

# Adding labels and title
plt.xlabel('solar_radiation')
plt.ylabel('Frequency')
plt.title('solar_radiation')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot for weather_temperature vs weather_humidity
plt.scatter(df['weather_temperature'], df['weather_humidity'], s=20)
plt.xlabel('Weather Temperature')
plt.ylabel('Weather Humidity')
plt.title('Weather Temperature vs Humidity')
plt.show()

# Calculate the correlation coefficient
correlation_coefficient = df['weather_temperature'].corr(df['weather_humidity'])

# Print correlation coefficient
print(correlation_coefficient)
```
